## Black Friday Dataset EDA and feature Engineering
### Cleaning and preparing a data for model training

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Import Dataset
df_train= pd.read_csv('Blackfriday_train.csv')
```

```python
df_test= pd.read_csv('Blackfriday_test.csv')
```

```python
df_train.head()
```

```python
df_test.head()
# Target feaature is not present in test data
```

```python
# TO perform Pre processing combine both datasets
df=df_train.append(df_test)
```

```python
df.head()
```

```python
## basic
df.info()
# Categorical var: 
```

```python
df.describe()
```

```python
# Drop non useful feature
df.drop(['User_ID'],axis=1,inplace=True)
# Axis:1 (Column wise),2(row wise)
# inpace = True(Delete entire column and update dataframe automatically)
```

```python
df.head()
```

```python
# handling a categorical features
# With the help of dummy var concept
df['Gender']=pd.get_dummies(df['Gender'],drop_first=1)
df.head()

# Process: convert categorical feture into numerical , append into original data set, delete categorical feature from dataframe
```

```python
# Map method(in original dataset itself categorical convert into Numerical(0,1))
# Gender
df['Gender']=df['Gender'].map({'F':0,'M':1})
df.head()
```

```python
df.head()
```

```python
# Age
# Find out unique values of column

df['Age'].unique()
```

```python
#pd.get_dummies(df['Age'],drop_first=True)
# Target Ordinal encoding : where order of diff categories is important
df['Age']=df['Age'].map({'0-17':1,'18-25':2,'26-35':3,'36-45':4,'46-50':5,'51-55':6,'55+':7})
```

```python
df.head()
```

```python
# City_Category
df_city = pd.get_dummies(df['City_Category'],drop_first=True)
df_city.head()
```

```python
df=pd.concat([df,df_city],axis=1)
df.head()
```

```python
df.drop(['B','C'],axis=1,inplace=True)
```

```python
# Drop City_Category feature
df.drop(['City_Category'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df=df.rename(columns={'B':'City_B','C':'City_C'})
```

```python
# Handle Missing Values
df.isnull().sum()
```

```python
# Focus on replacing missing values
df['Product_Category_2'].unique()
# This is a discrete feature(All this values gets repeated) 
```

```python
# Product_Category_2 handling missing values
df['Product_Category_2'].value_counts()
```

```python
# Replace missing value with Mode
df['Product_Category_2']=df['Product_Category_2'].fillna(df['Product_Category_2'].mode()[0])
```

```python
df['Product_Category_2'].mode()[0]
# Get series , so need to use indexing 
```

```python
# Cross check
df['Product_Category_2'].isnull().sum()
```

```python
# Product_Category_3 handling missing values
df['Product_Category_3'].value_counts()
```

```python
# Replace missing value with Mode
df['Product_Category_3']=df['Product_Category_3'].fillna(df['Product_Category_3'].mode()[0])
```

```python
# Cross check
df['Product_Category_2'].isnull().sum()
```

```python
df.head()
```

```python
df['Stay_In_Current_City_Years'].unique()
```

```python
df['Stay_In_Current_City_Years']=df['Stay_In_Current_City_Years'].str.replace('+',' ')
```

```python
df.head()
```

```python
df.info()
# Stay_In_Current_City_Years is Object , so need to conver into integer
```

```python
# Change datatype of Stay_In_Current_City_Years Featue (Object to Integer)
df['Stay_In_Current_City_Years']=df['Stay_In_Current_City_Years'].astype(int)
```

```python
df.info()
```

```python
df['Gender']=df['Gender'].astype(int)
```

```python
df.info()
```

```python
df['City_B']=df['City_B'].astype(int)
df['City_C']=df['City_C'].astype(int)
```

```python
df.info()
```
